```python
import numpy as np
import matplotlib.pyplot as plt

class RunningLog:
    
    runs = []

    def __init__(self):
        self.runs = []

    def add_run(self, hms, dist_km):
        self.runs.append([hms, dist_km])

    def num_runs(self):
        number = len(self.runs)
        return number

    def plot(self):
        """ Generate a line plot showing the average pace (minutes per kilometer) for
        run1, run2, .... run N. The x-axis is the run number, and the y-axis is the average
        pace for that run. """
        a = []
        c = []
        
        for i in range(self.num_runs()):
            a.append(i+1)
            b = self.runs[i][0][3:5] + "." + self.runs[i][0][6:8] 
            b = float(b)/self.runs[i][1]
            c.append(b)
        
        plt.plot(a,c)   
        plt.xticks(range(1,self.num_runs()+1))
        plt.xticks(rotation=90)
        plt.show()
       

    def save(self, filename):
        """ OPTIONAL FOR A FIVE: Save the data to a file. """

    def load(self, filename):
        """ OPTIONAL FOR A FIVE: Load the data from a file """
        with open(filename,'r') as f:
            next(f)
            for line in f:
                lines = line.split()
                lines[1] = "00:" + lines[1][0:5]
                self.add_run(lines[1],5)
                
```

```python
def main():

    # Instantiate a new running log
    logger = RunningLog()

    # This is optional
    logger.load("running.log")

    # Here are 4 sample running events
    logger.add_run("00:32:14", 5.1)
    logger.add_run("00:34:59", 5.5)
    logger.add_run("00:29:01", 4.9)
    logger.add_run("00:30:00", 5.0)
    logger.add_run("00:34:23", 5.3)
    logger.add_run("00:45:36", 7.5)
    logger.add_run("00:23:34", 3.8)
    logger.add_run("00:35:00", 5.7)
    logger.add_run("00:28:05", 4.7)
    logger.add_run("00:33:33", 5.3)

    # Output some data
    print(f"There are {logger.num_runs()} runs in the database")
    logger.plot()

    # This is also optional
    logger.save("running.log")
```

```python
if __name__ == '__main__':
    main()
```
