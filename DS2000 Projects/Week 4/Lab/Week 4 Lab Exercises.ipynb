```python
import pandas as pd
```

## Exercise 1

```python
#Reads in the csv file and quickly describes the data
pokemon = pd.read_csv("Pokemon.csv")
pokemon
```

```python
pokemon.describe()
```

## Execise 2

```python
#Strips the pokemon names of any unnecessary spaces and ensures they all start with a capital
pokemon['Name'] = pokemon['Name'].apply(lambda x: x.title().strip())
```

```python
#gets rid of any digits in the pokemon names
pokemon['Name'] = pokemon['Name'].replace('\d+', '', regex=True)
pokemon
```

```python
#gets rid of any special characters
pokemon['Name'] = pokemon['Name'].replace('[!,.,?, %]', '', regex= True)
pokemon
```

```python
#the drop_duplicates function drops any duplicates with the same pokedex number
pokemon.drop_duplicates(subset = ["#"], inplace = True)
pokemon
```

## Exercise 3

```python
#Creates a dictionary of pokemon, with the pokemon names serving as the keys
pokemondict = pokemon.set_index('Name').T.to_dict('list')
pokemondict
```

## Exercise 4

```python
#Asks the user to input 3 pokemon names
list = []
for i in range(3):
    list.append(input('Enter a Pokemon Name: ').title().strip())
```

```python
#checks if the inputting pokemon are in the pokedex
for i in list:
    if i in pokemondict and (pokemon==i).any().any():
        print(f"{i} is in the pokedex")
    else:
        print(f"{i} is not in the pokedex")
```

## Exercise 5

```python
#Creates a dataframe based on each type 
pokemonMeans = pokemon.groupby('Type 1').mean(numeric_only = True)
pokemonMeans
```

```python
#Creates a function that makes a 3d Array filled with the A,D,H mean values of inputted types
def typeMeans (list):
    values = []
    for i in list:
        values.append([ pokemonMeans.loc[i]['Attack'], pokemonMeans.loc[i]['Defense'], pokemonMeans.loc[i]['HP'] ] )
    print(values)
```

```python
#Test of function
list = ['Bug', 'Normal', 'Dragon']
typeMeans(list)
```

## Exercise 6

```python
#a function that creates an array of the ADH values of inputted pokemon names
def findPokemon(list):
    values = []
    for i in list:
        if i in pokemondict:
            values.append([i, pokemondict[i][4:7]])
    print(values)
```

```python
list = ['Charmander', "Ivysaur"]
```

```python
findPokemon(list)
```

## Exercise 7

```python
import math, time
```

```python
#Euclidian distance formula with a timer to check accuracy
def distance(list1, list2):
    start = time.time()
    d = math.sqrt( (list2[0]-list1[0])**2 + (list2[1]-list1[1])**2 + (list2[2]-list1[2])**2)
    end = time.time()
    print(d)
    #declares the total seconds for the function to run
    print(f"{end-start} seconds")
```

```python
list1 = [12,34, 67]
list2 = [42,28,92]
```

```python
distance(list1, list2)
```

```python
#adds an assertion within the function to check if the formula I created is the same as the math.distance function 
def distance(list1, list2):
    start = time.time()
    d = math.sqrt( (list2[0]-list1[0])**2 + (list2[1]-list1[1])**2 + (list2[2]-list1[2])**2)
    end = time.time()
    assert d==math.dist(list1,list2)
    print(d)
    print(f"{end-start} seconds")
```
