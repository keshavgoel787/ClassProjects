```python
import re
import random
```

```python
#Reading in the file
HAM = "hamilton_lyrics.txt"
```

```python
#Read Lyrics function with an input for a file and strips any whitespace
def read_lyrics(filename):
    all_lines = []
    with open(filename) as infile:
        while True:
            line = infile.readline()
            if not line:
                break
            if line.strip() == "":
                continue
            
            all_lines.append(line.strip().lower())

    return all_lines


#Cleans the file by getting rid of the [words] and any punctuation
def clean(file):
    #Checks if there is any [word] and removes it
    for i, j in enumerate(file):
        if re.fullmatch('\[+.+\]', j):
            file.remove(file[i])
    #replaces any punctuation with a space    
    for i, j in enumerate(file):
        w = re.sub('[^\w\s]', '', j)
        file[i] = w
        
#Function that creates a list of the first word in every line    
def firstlist(file):
    firsts = []
    for i in file:
        #splits the sentence and takes the first word
        firsts.append(i.split(" ", 1)[0])
    return firsts

#Function that creates a list of the last word in every line
def lastlist(file):
    lasts = []
    for i in file:
        #splits the sentence and takes the last word
        lasts.append(i.split(" ")[-1])
    return lasts

#Creates a list of ngrams of size 2
def ngrams_generator(list):
    words = []
    for i in list:
        words += i.split()
    
    ngrams = zip(*[words[i:] for i in range(2)])
    # Concatentate into ngrams 
    return [" ".join(ngram) for ngram in ngrams]

#Converts the ngram to a dictionary by creating a dictionary with the first word of every ngram serving as the key
def toDict(list):
    wordDict = dict()
    for i,j in enumerate(list):
        a, b = j.split()
        if a not in wordDict:
            wordDict[a] = [b]
        elif a in wordDict:
            wordDict[a].append(b)
    return wordDict

#Generates random lyrics  
def generate_lyric(begin, ending, ngrams):
    ''' Function: generate_lyric
        Parameters: list of begin-words, list of end-words,
                    dictionary of key = word, value = list of followed-by
        Returns: one line of Hamilton lyric
    '''
    sentence = ""
    curr_word = random.choice(begin)
    while True:
        sentence += " " + curr_word
        if curr_word in ending:
            break
        curr_word = random.choice(ngrams[curr_word])
    return sentence
```

```python
#Reads the lyrivs for Hamilton
hamilton = read_lyrics(HAM)
```

```python
#Cleans the file
clean(hamilton)
```

```python
#the list of first words
firsts = firstlist(hamilton)
```

```python
#the list of last words
lasts = lastlist(hamilton)
```

```python
#A list of ngrams
ngram = ngrams_generator(hamilton)
ngram
```

```python
#Converts the ngram list to a dictionary 
ngram = toDict(ngram)
ngram
```

```python
#generates random lyrics
for i in range(40):
    print(generate_lyric(firsts, lasts, ngram))
```
