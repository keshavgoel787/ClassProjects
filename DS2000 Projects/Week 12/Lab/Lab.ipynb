```python
from drv import DRV
```

```python
def main():
    R = DRV({1.5:.25, 2.0:.25, 2.5:.25, 3.0:.25})
    fp = DRV({.85:.5, .9:.25, .95:.25})
    ne = DRV({1:.2, 2:.2, 3:.2, 4:.2, 5:.2})
    f1 = DRV({1:1})
    fi = DRV({1:1})
    fc = DRV({.1:.5, .2:.5})
    L = DRV({10000: .2 , 100000: .2 , 1000000: .2, 10000000: .2, 100000000: .2})
    
    N = R*fp*ne*f1*fi*fc*L
    
    N.plot('Drake Equation Estimates', show_cumulative = False, 
          trials = 100000, yscale = 'log')
    
    print(N.E())
```

```python
main()
```
