## Exercise 1: Palindromes

```python
#asks the user to input a sentence
sentence = input("Enter a sentence: ")
```

```python
#Puts the sentence into a list, splitting by the spaces 
sentenceList = sentence.split(" ")
```

```python
#iterates through the list
for i in sentenceList:
   #sets a word2 variable that reverses i
    word2 = i[::-1]
    #check if i and word2 are equal, if they are then it is a palindrome, if not then it is not a palindrome
    if(word2 == i):
        print(f"{i} is a palindrome")
    else:
        print(f"{i} is not a palindrome")
        
```

## Exercise 2: Word Frequency Calculator

```python
#Sets variable to the quote by Thrun
thrunQuote = 'Nobody phrases it this way, but I think that artificial intelligence is almost a humanities discipline. It is really an attempt to understand human intelligence and human cognition'
#Sets all the words to a lower case to make it easier to allow duplicates
thrunQuote = thrunQuote.lower()
```

```python
#sets the thrunQuote to a list split by spaces and sets an empty dictionary
thrunList = thrunQuote.split(" ")
thrunDict = {}
```

```python
#iterates through the thrunList
for i in thrunList:
    #if the word is already in the dictionary then increase count by 1
    if i in thrunDict:
        thrunDict[i]+=1
    #if not then add the word to the dictionary with a value of 1 
    elif i not in thrunDict:
        thrunDict[i] = 1
```

## Exercise 3: Pangram Checker

```python
#sets a tuple of the alphabet, asks the user for a sentence, and sets a boolean for the result
alphabet = ("a", "b", "c", "d", "e", "f", "g", "h", "i", "j", "k", "l", "m", "n", "o", "p", "q", "r", "s", "t", "u", "v", "w", "x", "y", "z")
sentence = input("Enter a sentence: ").lower()
pangramResult = True
```

```python
#iterates through tuple of alphabet and checks if each letter is in the sentence
for _ in alphabet:
    if _ not in sentence:
        pangramResult = False
```

```python
#returns the result
if pangramResult:
    print("The sentence is a pangram")
else:
    print("The sentence is not a pangram")
```
